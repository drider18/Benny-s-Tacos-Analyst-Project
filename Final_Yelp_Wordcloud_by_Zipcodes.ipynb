```
#Import Packages required
from wordcloud import WordCloud
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
# read yelp_90013.csv into a DataFrame using a relative path
yelp_90013 = pd.read_csv('/content/drive/MyDrive/BSAN6060/yelp_review_90013.csv')

yelp_90014 = pd.read_csv('/content/drive/MyDrive/BSAN6060/yelp_review_90014.csv')

yelp_90046 = pd.read_csv('/content/drive/MyDrive/BSAN6060/yelp_review_90046.csv')

yelp_90232 = pd.read_csv('/content/drive/MyDrive/BSAN6060/yelp_review_90232.csv')

```

## Analysis on 90013 Zipcode

```
yelp_90013.head(4)
```

```
yelp_90013.tail(4)
```

```
yelp_90013.shape
```

```
yelp_90013.describe()
```

**Exploratory Data Analysis**

Below we visualize some of the reviews from the 90013 zipcode


```
import seaborn as sns

ax = sns.barplot(data=yelp_90013, x='review_rating', y='review_rating', estimator=lambda x: len(x) / len(yelp_90013) * 100)
ax.set(ylabel="Percent")
plt.title('Distribution of Customer Rating in 90013')
plt.show()
```

**Word Cloud**

Word clouds are especially effective for us because we are looking at review data.  Below you can see the various sentiments with and without stop words from the 90013 zipcode.

```
# concatenate all the reviews into one single string 
full_text = ' '.join(yelp_90013['text'])

my_stop_words = ["LA","go","place", "food", "spot", "grand", "central", "DTLA"] + list(STOPWORDS)
cloud_no_stopword = WordCloud(width=1600, height=800,max_font_size=200, background_color='white', stopwords=my_stop_words).generate(full_text)
plt.figure(figsize=(12,10))
plt.imshow(cloud_no_stopword, interpolation='bilinear')
plt.axis('off')
plt.show()
```
