```python
#import package to make HTTP requests (similar to a browser request)
import requests
import json
import pandas as pd
```

```python
yelp_rest = pd.DataFrame(None)

for offset_num in range(0, 1000, 50):
    
    
    
    #set params
    
    api_url = 'https://api.yelp.com/v3/businesses/search'
    params = {'term' : 'mexican', 'location' : 'Los Angeles', 'sort_by' : 'distance', 
             'offset': offset_num, 'limit': 50}

    #set key/headers
    
    api_key = 'lHNoGitPyFt2COwE_nMbGDQZM6pmgd0scwqiEDSiV22PKvVjt_eDZxHLJNN5rv1gmMFvlzN05dx0hxuXr47GM1iAB_gbsD4RlpBsvPC7myo4RfzSTwAiHDZYU-3mYXYx'
    headers = {'Authorization' : 'Bearer %s' % api_key}
    
    #make an API request
    
    api_request = requests.get(api_url, params = params, headers=headers) 
    
    #get response for a single page of 50
    
    json.loads(api_request.text)
    api_response = json.loads(api_request.text)
    businesses = api_response['businesses']

    
    business_data = {
    'name' : [],
    'review_count' : [],
    'rating':[],
    'address1':[],
    'address2':[],
    'city':[],
    'city':[],
    'state':[],
    'zip_code':[],
    'url':[],
    'id':[], 
    'latitude':[], 'longitude':[], 'alias':[], 'title':[], 'price':[]
    }



    for business in businesses:
        name = business['name']
        business_data['name'].append(name)
        #print(name)

        review_count = business['review_count']
        business_data['review_count'].append(review_count)
        #print(review_count)

        rating = business['rating']
        business_data['rating'].append(rating)
        #print(rating)

        address1 = business['location']['address1']
        business_data['address1'].append(address1)
        #print(address1)

        address2 = business['location']['address2']
        business_data['address2'].append(address2)
        #print(address2)

        city = business['location']['city']
        business_data['city'].append(city)
        #print(city)

        state = business['location']['state']
        business_data['state'].append(state)
        #print(state)

        zip_code = business['location']['zip_code']
        business_data['zip_code'].append(zip_code)
        #print(zip_code)

        url = business['url']
        business_data['url'].append(url)
        #print(url)

        id = business['id']
        business_data['id'].append(id)
        #print(id)
        
        latitude = business['coordinates']['latitude']
        business_data['latitude'].append(latitude)
        
        long = business['coordinates']['longitude']
        business_data['longitude'].append(long)
                                          
        alias = business['categories'][0]['alias']
        business_data['alias'].append(alias)
        
        title = business['categories'][0]['title']
        business_data['title'].append(title)
        
        
        def hello(num):
            return businesses[num]['price']
        
        for i in range (0,50,1):  
            try:
                hello(i)
            except Exception:
                price = 'NA'
            else:
                price = (businesses[i]['price'])
                
        business_data['price'].append(price)

        #print('-'*75)
        
        # convert Python dict to a pandas dataframe
        
        df = pd.DataFrame(business_data)
        
    yelp_rest = yelp_rest.append(df)
        
    print(offset_num)

```

```python
yelp_rest.to_csv('yelp_csv2.csv', index = False)
```
